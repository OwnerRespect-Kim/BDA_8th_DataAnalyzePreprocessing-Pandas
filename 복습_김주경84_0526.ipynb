### 데이터 시각화 (Data Visualization)
- 데이터 타입에 따른 시각화 방법을 다양하게 사용해서 내가 주장하고자 하는 분석 결과에 대해서 근거자료로 잘 사용할 수 있도록 해야한다.

```python
from matplotlib import pyplot as plt # 시각화 패키지
# %matplotlib inline : plt.show()를 하지 않아도 자동으로 시각화 코드가 나오게 하는 방법이다.
%matplotlib inline
```

```python
# scatter x, y축에 대한 값을 받아서 그림이 그려지게 한다.
plt.scatter([5, 6, 7, 8], [1, 2, 3, 4])
plt.title('x axis and y axis')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# 보고싶은 축의 범위를 조정하고 싶다.
# xlim, ylim을 사용하면 x, y축의 범위를 조정할 수 있다.
# ex
plt.scatter([5, 6, 7, 8], [1, 2, 3, 4])
plt.xlim(6, 8)
plt.ylim(2, 4)
plt.title('x axis and y axis')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# 보고싶은 축의 눈금값을 조정하고 싶다.
# xticks, yticks를 사용하면 x, y축의 눈금값을 조정할 수 있다.
# ex)
plt.scatter([5, 6, 7, 8], [1, 2, 3, 4])
plt.xticks([5, 6, 7, 8], ['V', 'VI', 'VII', 'VIII'])
plt.yticks([1, 2, 3, 4], ['I', 'II', 'III', 'IV'])
plt.title('x axis and y axis')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# 두 개의 그래프를 그리고 싶다.
# scatter를 두 번 사용하면 된다.
plt.scatter([3, 6, 9, 12], [15, 18, 21, 24], label = 'First')
plt.scatter([2, 4, 6, 8], [10, 12, 14, 16], label = 'Second')
plt.title('x axis and y axis')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 범례를 표시한다. loc = 'best'는 최적의 위치에 범례를 표시한다.
plt.legend(loc = 'best')
```

```python
# 도화지의 사이즈를 정하고 싶다.
# figure를 사용하면 도화지의 사이즈를 조정할 수 있다.
plt.figure(figsize = (15, 10))
plt.scatter([5, 6, 7, 8], [1, 2, 3, 4], label = 'Test')
plt.legend(loc = 'center')
```

```python
# 도화지의 스타일을 바꾸고 싶다.
# seaborn 확장 라이브러리를 사용하면 도화지의 스타일을 바꿀 수 있다.
import seaborn as sns
sns.set(style='whitegrid')
plt.figure(figsize=(10, 5))
plt.scatter([5, 6, 7, 8], [1, 2, 3, 4], label = 'Test')
```

```python
# .이 아니라 선으로 그래프를 그리고 싶다.
# plot을 사용하면 선으로 그래프를 그릴 수 있다.
plt.figure(figsize=(10,5))
plt.plot([1, 2, 3], [4, 5, 6], label ='Line1')
plt.plot([1, 2, 3], [6, 5, 4], label ='Line2')
plt.legend(loc = 'best')
```

```python
# plot 그래프에 대한 색상을 바꾸고 싶다.
# plot의 세 번째 인자로 색상을 지정할 수 있다. ex) color = 'red'
plt.figure(figsize=(10,5))
plt.plot([1, 2, 3], [4, 5, 6], label ='Line1', color = 'gray')
plt.plot([1, 2, 3], [6, 5, 4], label ='Line2', color = 'black')
plt.legend(loc = 'best')
```

```python
# plot 그래프에 대한 선의 모양을 바꾸고 싶다.
# plot의 네 번째 인자로 선의 모양을 지정할 수 있다. ex) linestyle = '--'
plt.figure(figsize=(10,5))
plt.plot([1, 2, 3], [4, 5, 6], label ='Line1', color = 'gray', linestyle = 'dotted')
plt.plot([1, 2, 3], [6, 5, 4], label ='Line2', color = 'black', linestyle = 'dashed')
plt.legend(loc = 'best')
```

```python
# plot 그래프에 마커를 표시하고 싶다.
# plot의 인자로 마커를 지정할 수 있다. ex) marker = 'o'
plt.figure(figsize=(10,5))
plt.plot([1, 2, 3], [4, 5, 6], label ='Line1', color = 'gray', linestyle = 'dotted', marker = 'o')
plt.plot([1, 2, 3], [6, 5, 4], label ='Line2', color = 'black', linestyle = 'dashed', marker = '*')
plt.legend(loc = 'best')
```

```python
# plot 그래프에 대한 투명도를 조절하고 싶다.
# plot의 인자로 투명도를 지정할 수 있다. ex) alpha = 0.5 (1이 최대, 0이 최소)
plt.figure(figsize=(10,5))
plt.plot([1, 2, 3], [4, 5, 6], label ='Line1', color = 'gray', linestyle = 'dotted', marker = 'o', alpha = 0.5)
plt.plot([1, 2, 3], [6, 5, 4], label ='Line2', color = 'black', linestyle = 'dashed', marker = '*', alpha = 0.05)
plt.legend(loc = 'best')
```

```python
# 히트맵을 그리고 싶다.
import pandas as pd

df = pd.DataFrame({'+score' : [5, 6, 7],
                   '-score' : [1, 2, 3]},
                   index = ['Korean', 'Math', 'English'])
```

```python
# sns 사용
# 더 높은 점수가 더 밝은 색으로 표시된다.
# 더 낮은 점수가 더 어두운 색으로 표시된다.
sns.heatmap(df)
```

```python
# 히트맵의 수치를 표시하고 싶다.
# annot = True를 사용하면 수치를 표시할 수 있다.
# fmt = 'd'를 사용하면 정수로 표시, fmt = '.1f'를 사용하면 소수점 첫째자리까지 표시
sns.heatmap(df, annot=True, fmt='.1f')
```

```python
# 히트맵의 색상을 바꾸고 싶다.
# cmap을 사용하면 색상을 바꿀 수 있다.
# linecolor를 사용하면 각 셀을 구분하는 선의 색상을 바꿀 수 있다.
sns.heatmap(df, annot=True, fmt='.1f', cmap='coolwarm', linecolor='white')
```

### 반복문을 이용해서 여러가지 그래프를 한 번에 그리기

```python
df_titanic = sns.load_dataset('titanic')
```

```python
df_titanic
```

```python
# 상관관계 계산을 위해 숫자형 데이터만 선택
corr_data = df_titanic.select_dtypes(include='number').corr()

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

plots = [
    {'func': sns.heatmap, 'data': corr_data, 'annot': True, 'title': 'Correlation Heatmap', 'ax': axs[0,0]},
    {'func': sns.violinplot, 'x': 'class', 'y': 'age', 'hue': 'survived', 'split': True, 'title': 'Age Distribution by Class and Survival', 'ax': axs[0,1]},
    {'func': sns.countplot, 'x': 'pclass', 'hue': 'survived', 'title': 'Survival Count by Class', 'ax': axs[1,0]},
    {'func': sns.boxplot, 'x': 'sex', 'y': 'fare', 'title': 'Fare Distribution by Sex', 'ax': axs[1,1]}
]

for plot in plots:
    if plot['func'] == sns.heatmap:
        plot['func'](data=plot['data'], annot=plot.get('annot', False), ax=plot['ax'])
    elif plot['func'] == sns.violinplot:
        plot['func'](data=df_titanic, x=plot.get('x'), y=plot.get('y', None), hue=plot.get('hue', None), split=plot.get('split', False), ax=plot['ax'])
    else:
        plot['func'](data=df_titanic, x=plot.get('x'), y=plot.get('y', None), hue=plot.get('hue', None), ax=plot['ax'])
    plot['ax'].set_title(plot['title'])

# 그래프 충돌 방지
plt.tight_layout()
plt.show()
```
