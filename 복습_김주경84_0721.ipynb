```python
import pandas as pd

# 샘플 DataFrame 생성
data = {
    'Name' : ['Kim', 'Na', 'Park', 'Lee', 'Ha'],
    'E-mail' : ['Kim@example.com', 'Na@sample.com', 'Park@case.com', 'Lee@standard.com', 'Ha@test.com'],
    'Explanation' : ['Kim is a Analyst.', 'Na is a Scientist.', 'Park is a Developer.', 'Lee is a Engineer.', 'Ha is a Designer.']
}
df = pd.DataFrame(data)
display(df)
```

```python
## cat을 이용하여 데이터 붙이기
df['Name+Email'] = df['Name'].str.cat(df['E-mail'], sep = ' / ')
```

```python
df
```

```python
## split을 이용하여 데이터 분리하기
df[['UserID', 'Domain']] = df ['E-mail'].str.split('@', expand = True)
```

```python
df
```

```python
## get을 이용하여 문자의 특정 위치의 문자를 가져오기
df['Explanation'].str.get(3)
```

```python
## slice를 이용하여 문자열의 일부분을 가져오기
display(df['Explanation'].str.slice(0, 5))
## start = 시작 숫자, end = 끝 숫자를 기입하여서도 가능하다. 
display(df['Explanation'].str.slice(start = 0, stop = 5))
```

```python
## extract를 이용하여 문자열에서 특정 문자를 추출하기
df['E-mail'].str.extract(r'@(\w+\.\w+)')
# 'r' : 'r' 접두사는 raw string을 의미하며 백슬래시를 문자 그대로 인식하게 한다.
# '@' : '@' 문자를 찾는다.
# '('와 ')' : 캡처 그룹을 정의하며, 캡처 그룹 안의 내용이 추출되게 한다.
# '\w+' : 하나 이상의 단어 문자를 찾으며, 단어문자는 알파벳(대소문자), 숫자, 밑줄(_)을 포함다.
# '\.' : '.' 문자 그대로의 마침표를 찾기 위해 '\.'을 사용한다.
```

```python
df
```

```python
## extractall을 사용하여 정규표현식에 일치하는 모든 부분을 추출하기
df['Explanation'].str.extractall(r'(\w+)')
```

```python
## replace를 사용하여 문자열 대체하기
df['Explanation'].str.replace('is', 'is not')
```

```python
## find를 사용하여 문자열에서 특정 문자의 위치를 찾기
df['Explanation'].str.find('Analyst')
```

```python
## contains를 사용하여 문자열이 특정 문자를 포함하는지 확인하기
## 반환값은 포함하면 True, 포함하지 않으면 False
df['Explanation'].str.contains('Analyst')
```

```python
## loc를 응용하여 특정 문자열을 포함하는 행을 추출하기
df.loc[df['Explanation'].str.contains('Analyst')]
```

### 원핫인코딩(One-Hot Encoding)이란?
- 각 범주형 변수에 대해 하나의 열을 생성
- 특정 행에 속한 범주에 해당하는 열에는 1을, 나머지 열에는 0을 넣는 것
- get_dummies() 함수를 이용하여 가능

```python
## get_dummies를 사용하여 문자열을 원핫인코딩하기
df['Domain'].str.get_dummies()
```

```python
df[['Domain', 'UserID']].str.get_dummies()
# 오류가 발생한다.
# str 속성은 1차원 데이터인 'Series' 객체에 대해 문자열 메서드를 사용할 수 있기 때문이다.
# 'DataFrame
# 객체는 2차원 데이터로 str 속성을 사용할 수 없다.
# pd.get_dummies() 함수를 사용하여 데이터프레임을 원핫인코딩할 수 있다.
```

```python
# pd.get_dummies() 함수를 사용하여 데이터프레임을 원핫인코딩할 수 있다.
pd.get_dummies(df[['Domain', 'UserID']])
```

### 텍스트 전처리

- 자연어처리(NLP) 인간의 언어 -> 컴퓨터가 이해할 수 있도록 처리한다. 즉, 컴퓨터가 분석할 수 있도록 프로그래밍 하는 방법

```python
import nltk
```

```python
#nltk 라이브러리 
nltk.download('punkt')
nltk.download('webtext')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

#### 전처리 단계  
- 텍스트에 대한 전처리 단계  
- 텍스트가 주어진다면 분석에 불필요한 것들이 분명히 존재할 것이다.  
  - 그러한 것들을 제거하고 문장들을 나눠서 분석할 수 있다.  
  - 표준 단어 같은 것들로 분리하고, 품사 등을 파악  
- 한국어와 영어는 다르기에 전처리 방식도 다를 것  
  
#### 전처리 용어 정리  
- 정제 : 정규표현식 등으로 데이터에 노이즈를 제거하거나, 필요한 것들만 추출하는 작업  
- 토큰화 : 주어진 텍스트를 원하는 단위로 나누는 것 (Token) 단어 토큰화, 다른 방식으로도 토큰을 나눌 수 있으며, 주로 딥러닝에서 성능을 올리기 위해 토큰화 작업을 진행  
- 정규화 : 어간 추출, 표제어 추출 등으로 작업  
- 품사 태킹 : 명사, 대명사, 형용사 등 앞서 토큰을 나누면 의미를 부여할 수 있다. 품사를 알려주면서 더 많은 의미있는 데이터를 확보한다.

```python
text_df = 'Deep learning is the subset of machine learning methods based on artificial neural networks with representation learning. The adjective "deep" refers to the use of multiple layers in the network. Methods used can be either supervised, semi-supervised or unsupervised.[2]'
```

```python
text_df
```

```python
## 문장에 대한 토큰화
from nltk.tokenize import sent_tokenize

display(sent_tokenize(text_df))
# 다양한 문장 구분 기호와 규칙을 사용하여 텍스트를 문장으로 분할한다.
```

```python
## 단어 토근화
from nltk.tokenize import word_tokenize

display(word_tokenize(text_df))
```

```python
## 특수문자 제거해주는 패키지
from nltk.tokenize import WordPunctTokenizer
# 특수문자 제거
display(WordPunctTokenizer().tokenize(text_df))
```

```python
## 정규표현식을 사용하여 단어 토큰화
from nltk.tokenize import RegexpTokenizer

regualr_tokenizer = RegexpTokenizer('[\w]+')
```

```python
display(regualr_tokenizer.tokenize(text_df))
```

```python
## 불용어 처리 (노이즈)
## 불용어 : 자주 등장하지만 분석에 필요하지 않은 단어(은, 는, 이, 가 등 의미가 없다고 보는 단어들)
## 영어도 불용어 처리를 진행해준다.
from nltk.corpus import stopwords # 불용어 사전

en_stops = set(stopwords.words('english'))
```

```python
print(len(en_stops))
print(en_stops)
```

```python
tokens = regualr_tokenizer.tokenize(text_df.lower())
tokens
```

```python
display([word for word in tokens if word not in en_stops]) # 불용어 처리가 된 것을 확인할 수 있다.
```
